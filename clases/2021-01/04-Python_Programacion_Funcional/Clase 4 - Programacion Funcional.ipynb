Una funci贸n es un segmento de c贸digo separado del c贸digo principal, la cual puede ser invocada desde cualquier otro punto del programa (incluyendo de si misma).

### Sintaxis

- En Python, una funci贸n se define por medio de la keyword ```def``` seguido por el nombre y los par谩metros que recibir谩 de entrada la funci贸n. (i.e., `def funcion(parametro_1, ...)`)

### Unit Testing


El *unit testing* o pruebas unitarias es un m茅todo para comprobar el correcto funcionamiento de un segmento de c贸digo o funci贸n.
La idea es crear casos de prueba en donde sabemos que valores deber铆a retornar una funci贸n y luego comprobar que la funci贸n efectivamente los retorne. 

Para esto, los test que creemos deben ser determin铆sticos (no aleatorios) y repetibles. 


Python provee la *keyword* `assert` la cual verifica el valor de una condici贸n: 

- Si es `True` continua la ejecuci贸n. 
- Si es `False`, lanza la excepci贸n `AssertionError` y detiene la ejecuci贸n.

````python
def suma(a, b):
    return a+b
````

````python
def suma(a, b):
    return a+b

assert suma(2,3) == 5
````

````python
# La idea es hacer varios casos de prueba unitarios.
assert suma(2,-3) == -1
assert suma(99999,1) == 100000
assert suma(3,3) != 5
````

Si por ejemplo, ahora modifico la funci贸n suma y multiplico a y b, el test deber铆a fallar.

````python
def suma(a, b):
    return a*b

assert suma(2,3) == 5
````

````python
# Le podemos indicar que nos entregue un mensaje de error

assert suma(3,2) == 5, 'Error al sumar'
````

#### Map, Filter y Reduce

En ciencia de datos, la utilidad de las funciones ```lambda``` generalmente se asocia a las operaciones ```map()```, ```filter()``` y ```reduce()``` usando usando el modulo functools). Est谩s operaciones se denotan como **funciones de orden superior** pues reciben otra funci贸n como argumento. 

**Map**

```map()``` permite aplicar la funci贸n objetivo sobre un contenedor (como una lista) elemento por elemento, el resultado es un objeto tipo ```map``` que entre sus caracter铆sticas es un iterable.


````python
# Ejemplo iterativo usando for
l_r = []
for elemento in range(1, 6):
    l_r.append(elemento ** 2)
l_r
````

````python
elevar_al_cuadrado = lambda x: x**2
lista = [1, 2, 3, 4, 5, 6]

map(elevar_al_cuadrado, lista) # Retorna un iterable
````

````python
 # Retorna un iterable. Para evaluarlo, usar list(map)
list(map(elevar_al_cuadrado, lista))
````

Por lo general, la funci贸n lambda se define dentro del map:

````python
lista = [1, 2, 3, 4, 5, 6]

list(map(lambda x: x**2, lista))
````

**Filter**

```filter()``` permite mantener elementos de un arreglo seg煤n el valor de verdad asociado a cada uno por la funci贸n objetivo.


````python
lista = [1, 2, 3, 4, 5, 6]

# Mantener solo numeros pares
list(filter(lambda x: x % 2 == 0, lista))
````

**Reduce**

`reduce()` permite acumular valores de izquierda a derecha seg煤n una funci贸n sobre alg煤n iterable. La idea es reducir la lista a un solo valor seg煤n la funci贸n estipulada.
El primer argumento de los 2 de la funci贸n a pasar es el valor acumulado y el segundo es el valor siguiente de la secuencia.

Nota: Hay que importarla desde `functools`

Por ejemplo:

```python
fun = lambda x, y: x + y
functools.reduce(fun, [1, 2, ,3, 4, 5])
```

Al ejecutar esta l铆nea se calcula:
1. fun(1, 2)
2. fun(fun(1, 2), 3) 
3. fun(fun(fun(1, 2), 3), 4)
4. fun(fun(fun(fun(1, 2), 3), 4), 5)

Lo que en total entrega,

$$\left(\left(\left(\left(1+2\right)+3\right)+4\right)+5 \right) = 15$$

Lo que en total entrega,

$$\left(\left(\left(\left(1+2\right)+3\right)+4\right)+5 \right) = 15$$


`functools.reduce` tambi茅n tiene un tercer (opcional) argumento, `initializer`, si este es entregado se utiliza como primer elemento en el calculo acumulativo, por lo que si en el caso anterior se tuviera

`functools.reduce(fun, [1, 2, ,3, 4, 5], 8)`

el resultado ser铆a,

$$8 + ((((1+2)+3)+4)+5) = 23$$

````python
from functools import reduce

lista = [1, 2, 3, 4, 5, 6]

# Sumar todos los elementos
reduce(lambda a, b: a + b, lista)
````

````python
# Encontrar el m谩ximo
reduce(lambda a, b: a if a > b else b, lista)
````

A continuaci贸n veremos algunos lineamientos a la hora de construir docstrings:

Para Docstrings de una linea:

* Se usa ```""" """``` inclusive si se puede escribir todo en una linea.
* Las comillas que inicial la documentaci贸n est谩n en la misma linea que aquellas que la cierran.
* El docstring es una frase que termina en punto, describe el objeto al cual se hace referencia y su efecto en la forma (accion,resultado).
* La documentaci贸n no debe tener la "firma" (signature) del objeto subyacente: 

```python
# Mala practica:

    """ funcion(a,b) -> int """
    return a+b
    
# Buena practica:

    """Suma (`a`+`b`) retorna un entero """
    return a+b
```

Docstrings multilinea:

* La documentaci贸n debe estar indentada completamente.

* La primera linea debe ser siempre un resumen corto y conciso de el prop贸sito del objeto que se documenta.

* Debe haber una linea en blanco luego del resumen corto. Se puede agregar una explicaci贸n m谩s profunda posterior al espacio.

En general existen cualidades comunes al momento de crear un docstring, estas incluyen, argumentos, atributos y resultados (returns). Los distintos est谩ndares de creaci贸n de documentos abordan esto, dentro de los est谩ndares m谩s comunes se encuentran:
