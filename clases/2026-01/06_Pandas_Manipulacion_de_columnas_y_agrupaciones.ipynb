Los bins resultantes pueden ser interpretados como categorías.

### 1.3 Dividir datos numéricos en cuantiles 

La función `pd.qcut` permite dividir los datos por cuantiles, los cuales a diferencia del anterior, tienen (_o intentan tener_) la misma cantidad de datos por cada intervalo generado.


Al igual que el caso anterior, podemos especificar como nombrar los cuantiles a través del atributo `labels`:

````python
air_pollution_qcut = pd.qcut(
    df_ambiental.loc[:, "Air pollution"],
    q=4,
)

air_pollution_qcut
````

````python
water_quality_qcut = pd.qcut(
    df_ambiental["Water quality"],
    q=5,
)

water_quality_qcut
````

### 1.4 Transformaciones personalizadas

También podemos aplicar funciones personalizadas que nosotros podemos definir. Esto permite aplicar funciones matemáticas, discretizaciones con bins no uniformes, agrupaciones, etc. Esto le da mucho más flexibilidad a la manipulación de columnas! 

````python
df_ambiental["Rooms per person"]
````

````python
df_ambiental["Rooms per person"].describe()
````

````python
df_ambiental["Rooms per person"].isna()
````

````python
df_ambiental["Rooms per person"].isna().sum()
````

`apply()` permite aplicar una función arbitraria a cada fila (`axis=1`) o columna (`axis=0`) de un DataFrame, o a cada elemento de una Serie.

> **⚠️ Advertencia de rendimiento**: `apply()` ejecuta la función en un bucle Python puro, por lo que es **significativamente más lento** que las operaciones vectorizadas nativas de pandas (como la aritmética entre columnas o `.str.method()`). Úsalo solo cuando no exista una alternativa vectorizada.


````python
def crowding_index(rooms_per_person: int | pd.nan) -> int:
    if not pd.isna(rooms_per_person):
        return (rooms_per_person - 1) * 2
    else:
        return 0
````

````python
crowding_index_col = df_ambiental["Rooms per person"].apply(crowding_index)
crowding_index_col
````

### 1.5 Dividir datos categóricos

Veamos otro ejemplo de funciones personalizadas, esta vez para agruár datos categóricos pero que están presentes en el índice. Esto presenta varios desafíos, ya que un índice no tiene las mismas propiedades que una columna

````python
paises_latinoamericanos = ["Chile",  "Brazil", "Colombia", "Mexico"]
df_ambiental.loc[paises_latinoamericanos]
````

````python
df_ambiental.index in paises_latinoamericanos
````

````python
def es_latino(pais):
    if pais in paises_latinoamericanos:
        return True
    return False
````

````python
df_ambiental.index.apply(es_latino)
````

`.apply()` no aplica para índices. ¿Qué podemos hacer al respecto?

Podríamos, por ejemplo, transformar el índice en columna

````python
df_ambiental.reset_index().head()
````

````python
es_latino_col = df_ambiental.reset_index()["Country"].apply(es_latino)
es_latino_col.head()
````

Pero ¿Qué pasó con el índice? Esto ocurrió porque para obtener esta columna tuvimos que resetear el índice de la tabla

````python
es_latino_col.index
````

> Podemos asignar directamente esta serie a una columna de df_ambiental? Por qué?

````python
# Guardamos para deshacer los cambios
df_ambiental["Es Latino"] = es_latino_col
df_ambiental.head()
````

````python
df_ambiental = df_ambiental.drop(columns=["Es Latino"])
df_ambiental.head()
````

Para asignar esto a una nueva columna deberíamos extraer los datos 'crudos' del DataSeries. Recordemos que también podemos asignar columnas con listas. Esto es:

```python
data_frame["Column"] = [value_1, value_2, ..., value_n]
```

En este caso, son datos sin índice ni información adicional, por lo que si son del mismo tamaño que la cantidad de filas de la tabla simplemente los asigna posicionalmente. Solo cuando existe información sobre índice pandas intenta utilizarlo para asignar qué valor va en qué fila. En python, se pueden extraer los datos crudos de un DataSeries utilizando `.values`. Es decir:

```python
data_frame["Column"] = data_series.values
```

O alternativamente para este caso de uso, podríamos usar `.isin()` directamente en el índice.

````python
es_latino_col= df_ambiental.index.isin(paises_latinoamericanos)
es_latino_col
````

````python
df_ambiental["Es Latino"] = es_latino_col
df_ambiental.head()
````

### 1.6 Asignación de nuevas columnas a un DataFrame

Hasta el momento hemos generado nuevas columnas, pero no las hemos guardado en ningún dataframe. Sin embargo sería idea poder unir estas columnas con sus respectivos datos originales.

**Si los índices del dataframe son los mismos que el de la nueva columna**, podemos asignar una serie como una nueva columna de nuestro DataFrame de la siguiente forma:

````python
df_ambiental.loc[:, "Air pollution Ranking"] = air_pollution_rank
df_ambiental
````

````python
df_ambiental.loc[:, "Air pollution Ranking"] = air_pollution_rank
df_ambiental.loc[:, "Air pollution Bins"] = air_pollution_cut
df_ambiental.loc[:, "Air pollution Quintile"] = air_pollution_qcut
df_ambiental.loc[:, "Crowding Index"] = crowding_index_col
df_ambiental.loc[:, "Es Latino"] = es_latino_col

df_ambiental.sort_values(by="Air pollution Ranking")
````

#### Valores Inválidos

Puede ocurrir que tengamos algún dato inválido que no podamos transformar a `datetime`.

`pd.to_datetime` ofrece el siguiente parámetro para manejar estos problemas:

`errors{‘ignore’, ‘raise’, ‘coerce’}, default ‘raise’`

- If `raise`, then invalid parsing will raise an exception.

### `agg`

`agg` permite agregar datos por grupo usando una o más operaciones a través de un diccionario.
El resultado también será un dataframe multi-índice, pero por columnas:

````python
stats_general.head()
````

````python
stats_general[stats_general["Temperature"] == stats_general["Temperature"].min()]#.min()
````

````python
stats_temperatura_globales = temperaturas.groupby(["Country", "Year"]).agg(
    {
        "Temperature": ["mean", "min", "max"]
    }
)
stats_temperatura_globales
````

````python
stats_temperatura_globales.columns
````

````python
stats_temperatura_globales
````

Al igual que el caso de los multi-índices en los índices de las filas, podemos acceder a ciertas series con multi-indices en las columnas a partir de indexadores basados en tuplas.

````python
stats_temperatura_globales.loc[:, ("Temperature", "mean")]
````

````python
stats_temperatura_globales.loc["Chile", ("Temperature", "mean")]
````

````python
stats_temperatura_globales.reset_index()
````

````python
stats_temperatura_años = stats_temperatura_globales.reset_index().groupby("Year").mean(numeric_only=True)
stats_temperatura_años
````
