## Ambientes virtuales y UV

### Ambientes virtuales

Por defecto, python tiene una carpeta global para paquetes. Cada vez que ejecutas ```pip install```, se agrega una librería en esta carpeta, que luego puedes usar en cualquier proyecto.

**El problema con esto**

Imaginemos que tienes 2 proyectos de ciencia de datos en tu computador:
- El **Proyecto A** usa una versión antigua de pandas (v 1.5) porque el proyecto fue hecho hace tiempo y utiliza una función de esta versión de pandas que luego fue deprecada
- El **Proyecto B** usa la versión más nueva de pandas (v 2.2) ya que fue desarrollado hace poco y utiliza las últimas funciones de pandas

Como solo existe una carpeta global, solo puedes tener **una** versión de pandas al mismo tiempo. Si actualizas pandas para trabajar en el proyecto B, se rompe la instalación para el proyecto A.

**Solución: Ambientes virtuales**

Un ambiente virtual es una carpeta local (usualmente llamada ```.venv```) dentro de tu proyecto que contiene las librerías usadas en el proyecto. Ventajas de esto:
- **Aislamiento:** El proyecto A tiene su propio ```.venv``` con pandas 1.5, y el proyecto B tiene su propio ```.venv``` con pandas 2.2. Nunca interfieren entre ellos
- **Limpieza:** Si quieres eliminar un proyecto, simplemente eliminas la carpeta y con esto se elimina el ambiente virtual, haciendo que librerías pesadas tambien se eliminen
- **Reproducibilidad:** Los softwares de ambientes virtuales usualmente tienen archivos que contienen todo lo necesario para reproducir las librerías usadas (requirements.txt o uv.lock)

> En softwares de ambientes virtuales como **anaconda**, los ambientes son independientes de los proyectos y no se almacenan en la carpeta de cada proyecto, sino que en los archivos de anaconda. Por esto, se pueden tener ambientes para diferentes propósitos y diferentes proyectos pueden usar el mismo ambiente. Sin embargo, en la práctica se tiende a tener un ambiente virtual por proyecto

Los ambientes virtuales no son solo una 'buena práctica', sino que una **táctica escencial** para desarrolladores

### UV

<div align="center">


</div>

UV es un manager de paquetes y proyectos escrito en **Rust** para python.
- Es una **herramienta** externa _standalone_, no una librería.
- UV puede instalar a python. Puede ir "antes" de python
- Remplaza a **pip** en manejo de paquetes
- Remplaza a pyenv, conda y venv en manejar ambientes virtuales
- Es hasta **100** veces más rápido que pip:
    - Maneja un caché global usado por cada ambiente, indicando las modificaciones
    - Tiene paralelización de tareas
- Remplaza el requirements.txt y por pyproject.toml y uv.lock

**Linux**

Ejecutar en una terminal

### Creando ambientes con UV

- Inicializamos la configuración del ambiente

```powershell
uv init --lib
```

Esto inicializa la configuración de nuestro nuevo ambiente virtual. Ejecutarlo creará un archivo [pyproject.toml](../../pyproject.toml) donde se encuentra esta configuración. Este archivo lo podemos mirar y modificar a mano, o modificar con comandos que nos aseguran que se mantenga la integridad del ambiente virtual


- Creamos un ambiente

```powershell
uv venv
```

O para crearlo con una versión específica de python
```powershell 
uv venv --python 3.14
```
Esto **crea** oficialmente nuestro ambiente virtual, que podemos notar por la aparición del directorio [.venv](../../.venv). Si bien hemos creado un ambiente, este todavía está vacío.

- Agregamos librerías

```powershell
uv add pandas ruff
```

Esto agrega las librerías pandas y ruff a nuestro ambiente virtual. Podemos notar la aparición del archivo [uv.lock](../../uv.lock). En este se "fija" el ambiente virtual al explicitar toda la configuración necesaria para reproducir en exactitud el ambiente ste se creó ya que intentamos instalar librerías sin un ambiente existente (previamente solo configuramos el ambiente). También podemos notar que el archivo [pyproject.toml](../../pyproject.toml) ha cambiado

> Nota: Si el ambiente virtual no existía al ejecutar esto, lo crea en el proceso

> Es posible que de error al agregar paquetes porque 'espera un __init__.py'. En ese caso, agregar lo siguiente a pyproject.toml:
> ```toml
> [tool.uv]
> managed = true
> package = false
> ```


- Activar el ambiente

Qué pasa ahora si intentamos ejecutar pandas?

```powershell
python
```
```python
import pandas
```

Obtenemos un ```ModuleNotFoundError: No module named 'pandas'```. ¿Por qué sucede? Porque todavía no estamos 'dentro' del ambiente. Para activar el ambiente y utilizar las librerías y versiones que definimos, debemos ejecutar
```powershell
.\.venv\Scripts\activate
```

Es tedioso tener que escribir '.\.venv\Scripts\activate' cada vez que quieres activar el ambiente. Por esto, recomiendo guardar ese comando en un archivo ```activate.bat```. Esto se puede hacer facilmente ejecutando

```powershell
echo .\.venv\Scripts\activate > activate.bat
```

Luego, para activar el ambiente basta con escribir
```powershell
activate
```

Para salir del ambiente virtual
```powershell
deactivate
```


- Sincronizar librerías
```powershell
uv sync
```

Este comando sincroniza el ambiente virtual con nuevos cambios existentes en [pyproject.toml](../../pyproject.toml). Por ejemplo, podemos abrir este archivo y agregar matplotlib como dependencia. Luego de esto ejecutamos ```uv sync``` y se instalará matplotlib.

> También podemos utilizar este comando cuando estemos trabajando en un repositorio compartido por muchos colaboradores. Si alguno de ellos cambia la versión de alguna librería y por ende al ejecutar el programa hay errores ```uv sync``` debería arreglarlo.

#### Importancia de pyproject.toml

Previamente a UV, para agregar librerías se tenía que editar un archivo requirements.txt, mientras que la configuración se ubicaba en archivos como runtime.txt, setup.py y archivos de flake8. pyproject.toml centraliza todo eso como una singular fuente de la verdad.

Incluso es posible configurar ruff en este archivo:

```toml
# Agregar esto al final de pyproject.toml
[tool.ruff]
line-length = 100

[tool.ruff.lint]
select = ["E", "F", "I"]

[tool.ruff.format]
# This makes Ruff's formatter behave exactly like Black
quote-style = "double"
indent-style = "space"
```

## Ruff

Ruff es un linter y formateador de alta eficiencia escrito en Rust. En el pasado, los desarrolladores de python tenían que combinar muchas herramientas para tener código bien formateado, como Flake8 para encontrar errores, isort para ordenar imports, y Black para formatear código.Ruff reemplaza todas estas herramientas en una sola, y es 10 - 100 veces más rápido.

En las secciones anteriores, instalamos la librería ruff en nuestro ambiente virtual con ```uv```, y también instalamos la extensión de VS Code. Debido a la extensión y la librería, es posible ejecutar Ruff automáticamente al guardar un script. Para esto, se debe editar la configuración de VS Code

**IMPORTANTE:** Se exigirá que los entregables de los laboratorios hayan pasado por este tipo de formateo. Archivos mal formateados son realmente difíciles de leer, y conllevará descuentos

Para ejecutar ruff al guardar archivos:

- Ir al panel de comandos ```ctrl + shift + P```
- Buscar User Settings (JSON)
- Agregar los siguiente y consolidar (Eliminar llaves duplicadas si las hay)

```json
{
    "editor.defaultFormatter": "charliermarsh.ruff",
    "editor.formatOnSave": true,
    "editor.codeActionsOnSave": {
        "source.fixAll.ruff": "explicit",
        "source.organizeImports.ruff": "explicit"
    },

    "[python]": {
        "editor.defaultFormatter": "charliermarsh.ruff",
        "editor.formatOnSave": true,
        "diffEditor.ignoreTrimWhitespace": false,
        "editor.codeActionsOnSave": {
        "source.fixAll.ruff": "explicit",
        "source.organizeImports.ruff": "explicit"
        }
    },

    // For jupyter support
    "notebook.formatOnSave": true,
    "notebook.codeActionsOnSave": {
        "source.fixAll.ruff": "explicit",
        "source.organizeImports.ruff": "explicit"
    },

    // Ensures VS Code uses the Python environment created by 'uv'
    "python.defaultInterpreterPath": "${workspaceFolder}/.venv/bin/python",
    "python.terminal.activateEnvInSelectedTerminal": true,

  // To view the max line length
    "editor.rulers": [
        100
    ]
}
```

Esto permite que ruff se ejecute y formatee los archivos al guardarlos. Podemos ponerlo a prueba creando un archivo ```test.py``` con el siguiente código, y luego guardarlo con ```ctrl + s```

````python
lista_con_elementos = [1, 2, 3]


def mi_funcion_con_muchos_argumentos(a, b, c):
    return a + b + c

for elemento_en_lista in lista_con_elementos:
    print(mi_funcion_con_muchos_argumentos(elemento_en_lista, elemento_en_lista + 1, elemento_en_lista + 2))
````

## Git

### Control de Versiones con Git 🔃

Un **sistema de control de versiones (VCS)** es una herramienta que permite registrar cambios realizados en carpetas y archivos de un proyecto. Con tal registro es posible:

- Regresar a un estado anterior para deshacer errores.
- Colaborar de manera sincronizada a través de la nube.

Al trabajar en equipo, un sistema de control de versiones ayuda a identificar:

- Qué cambios fueron hechos.
- Quién/nes hicieron los cambios.
- Cuándo fueron hechos.

**`Git`** es una herramienta de este tipo y es la que usaremos en el curso. Es software libre y lo pueden encontrar en https://git-scm.com/.

### Ciclo de Vida de los Archivos 🗄️


`Git` permite llevar un registro de los cambios del proyecto por medio de la asignación de ciclos de vida a los archivos.

Visto de manera sencilla, se puede decir que en `git` existen 4 tipos de archivos:

- no rastreados o `untracked`.
- no modificados o `unmodified`.
- modificados o `modified`.
- listos para ser consilidados o `staged`.

Las interacciones son las siguientes:

#### Archivo consolidado

Un archivo consolidado o `commited`, es aquel que ha sido modificado y ha sido aprobado para formar parte integra del repositorio y del registro de cambios. 

> **Pregunta ❓**: ¿Este archivo ya se encuentra en el historial del repositorio?¿Y en el repositorio central?

#### Staging Area

La staging area consiste en el conjunto de archivos modificados que aún no han sido aprobados para formar parte del repositorio ni de su registro de cambios, pero que están a un paso de serlos.


Para que un archivo que ha sido modificado (`modified`) pase a ser parte de la staging area, debe estar *marcado* como `staged`. Esto puede ser realizado a través de `git add`:

```bash 
# Caso de un archivo:
git add path/to/file.py

# Caso de una carpeta:
git add path/*

# Caso de todos los cambios:
git add --all
```

Si agregamos `src/new_class.py` ahora ejecutamos `git status`, veremos algo como esto:

- Finalmente un archivo del repositorio puede no ser marcado como tracked, por lo que al modificarse, no pasa a ser parte del área de montaje. Este tipo de archivos de denota como **untracked**. 

#### Commit

Realizar un commit permite es capturar una instantanea de los archivos `staged` del repositorio y agregarlos a su historial de cambios.

Para ello utilizo:

```bash
git commit
```

Esto abrira un archivo en su editor de texto prederminado donde deberán dejar un mensaje de que cambios hicimos y porque. 

Si desean dejar una descripción pequeña de los cambios, simplemente podemos agregar el parámetro `-m`

```bash
git commit -m "Mi mensaje aquí"
```

Hacer un commit hará que los archivos `staged` pasen a ser `unmodified`.

Para observar el rgistro de cambios se utiliza ```git log```, al ejecutar este comando se aprecia la estructura de un commit: este consta de un autor, momento de realización de la modificación y un **hash** identificador del cambio.

### Deshacer Cambios ⬅️

#### Eliminar del area de montaje

¿Qué ocurre si se agrega accidentalmente un archivo al área de montaje? 

```bash
git reset HEAD
```

Nota: esto eliminará todos los archivos con cambios que hayan sido agregados al area de montaje.

#### Deshacer cambios de un archivo:

Si por otra parte, un archivo ```folder/file``` ya fue consolidado y se desean revertir los cambios al commit anterior, es posible usar el comando 

```bash
git checkout -- folder/file
```


Al igual que ```diff```, el comando ```checkout``` puede realizar múltiples funciones. Es posible volver a una versión (commit) anterior de cierto archivo utilizando la sintaxis:

```bash
git checkout commit-hash folder/file
```

Y si quieres retornar todos los archivos a un commit anterior, puedes usar:
    
```bash
git checkout commit-hash
```

### Branches

Una rama o **branch** es un mecanismo que nos permite trabajar en un ambiente independiente del código principal.

Al crear una branch, se "replican" los elementos de la rama desde donde se originó. Todos los commits que hagamos se harán en la branch que creamos. Luego, podemos juntar los cambios a través de un `merge`.

```bash
git branch new-branch
``` 

permite crear una rama con nombre *new-branch*. 

**Cambiar de ramas**

Para cambiar a la nueva rama, hacemos uso de 

```bash
git checkout new-branch
```

Para cambiar en general a cualquier rama:

```bash
git checkout <nombre-rama>
```

Un atajo para crear una nueva rama y acceder inmediatamente es el comando: 

```bash 
git checkout -b new-branch
```

### Merge

La gran ventaja de trabajar con ramas viene de unir posteriormente los resultados y registros, esto se denomina **merging**. Para unir dos ramas se utiliza el comando:

```bash
git merge source_branch destination_branch
```

### Conflictos 😥

Finalmente, es posible que existan colisiones en el trabajo realizado dentro de distintas ramas.
Estas ocurren cuando 2 personas trabajan en las mismas lineas de código y luego intentan unir sus cambios. Estas colisiones se denominan **conflictos** .

Comunmente uno deberá resolver estos conflictos a mano y luego subir los cambios.

### Repositorios Remotos 📡

Ya sabemos iniciar repositorios y clonarlos. 
En el segundo caso, `git` registra el repositorio original del cual se obtuvo un clon, tal repositorio original se denota como **remoto**. Tal información se almacena con un nombre y una dirección (local o url). 

El comando ```git clone ruta``` permite nombrar el repositorio clonado por medio de ```git clone ruta nombre_clon```. 


La ventaja de este sistema de almacenamiento de rutas remotas, es que permite sincronizar el trabajo tanto en una máquina local, como en colaboraciones por medio de internet. 

#### Pull

La comunicación entre repositorios se hace por medio de intrucciones especiales, una de ellas es ```pull```. 

```git pull``` permite obtener el registro de cambios de un repositorio remoto, e incluso, una rama de tal repositorio. La sintaxis es:

```bash
git pull repo_remoto rama
```

Con este comando, se toma toda la información del repositorio ```repo_remoto``` y se une directamente con el repositorio actual (*merging*).

#### Push

Por otra parte, es posible enviar cambios a un repositorio remoto, esto se hace por medio de ```git push```. La sintaxis es análoga: 

```bash
git push repo_remoto rama
```
