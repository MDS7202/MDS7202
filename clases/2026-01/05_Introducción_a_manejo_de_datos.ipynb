````python
nuevos_datos = np.concatenate([pais[:, np.newaxis], datos], axis=1)
nuevos_datos
````

> **Pregunta ❓**: ¿Qué `dtype` tiene ahora `nuevos_datos` después de concatenar strings y floats? Revisa `nuevos_datos.dtype`. ¿Por qué ocurre eso y qué consecuencias tiene para las operaciones numéricas?



````python
nuevos_datos.dtype
````

**Respuesta:** El resultado es `dtype('<U32')`, que significa *Unicode string de hasta 32 caracteres*.

Esto ocurre porque **NumPy solo puede almacenar un único tipo por array** (*homogeneidad*). Al concatenar floats con strings, NumPy no puede convertir los strings a número, pero sí puede convertir los números a texto. Entonces aplica la única conversión segura posible: todo se transforma en string.

El `32` indica el largo máximo reservado para cada elemento (en caracteres Unicode). NumPy lo calcula automáticamente según el string más largo del array.



````python
nuevos_datos.mean(axis=1)
````

**Consecuencias para operaciones numéricas:** ya no se puede operar aritméticamente con el array:
```python
nuevos_datos + 1   # → TypeError: ufunc 'add' did not contain a loop...
nuevos_datos.mean()  # → TypeError
```
Para recuperar los valores numéricos habría que convertir explícitamente: `nuevos_datos.astype(float)`, lo que fallaría si algún elemento es un string no numérico.

### `df["col"]` vs `df.loc[:, "col"]` y el `SettingWithCopyWarning`

Para **leer** datos ambas formas son equivalentes:

```python
df["Water quality"]        # ✅ lectura OK
df.loc[:, "Water quality"] # ✅ lectura OK
```

La diferencia importa al **escribir**. Considera el siguiente caso:

```python
df["Water quality"][0] = 99   # ⚠️ puede no modificar el DataFrame original
df.loc["Australia", "Water quality"] = 99  # ✅ siempre funciona correctamente
```

Cuando encadenamos dos indexadores (`df["col"][fila]`), pandas no puede garantizar si está modificando el DataFrame original o una copia temporal, y lanza un `SettingWithCopyWarning`.

> **Regla de oro 📌**: para **asignar** valores, usa siempre `df.loc[filas, columnas] = valor`.

> **Nota sobre pandas moderno 📖**: A partir de pandas 2.0 se introdujo _Copy-on-Write_ (CoW), que hace que cualquier slice sea siempre una copia independiente. Esto elimina el `SettingWithCopyWarning`, pero la práctica de usar `.loc` sigue siendo la más explícita y segura.


### Mascaras Booleanas y Consultas 🎭: Selección por Booleanos


Una operación interesante de selcción es usar un arreglo de booleanos para seleccionar datos.

````python
df
````

> **Pregunta ❓**: ¿Cómo podríamos obtener aquellos países cuya 80% o más de su población estén conformes con la calidad del agua?

````python
mask = df.loc[:, "Water quality"] >= 80
mask
````

````python
mask.sum()
````

````python
mask = df.loc[:, "Water quality"] == 92
# con .values -> numpy array
mask.sum()
````

````python
mask
````

````python
~mask
````

````python
df.loc[~mask, "Water quality"]
````

````python
mask
````

````python
mascara = df["Water quality"] >= 101
mascara
````

````python
df.loc[mascara, ["Water quality"]]
````

> **Nota 📖**: Como `Water quality` representa un porcentaje (0–100), ningún valor puede superar 101. La máscara es siempre `False` y el resultado es un DataFrame vacío. Esto ocurre en la práctica cuando los límites de un filtro no corresponden al rango real de los datos.


````python
boolean_mask = (df.loc[:, "Water quality"] > 80) | (df.loc[:, "Water quality"] < 94)
df.loc[boolean_mask]
````

````python
boolean_mask.sum()
````

````python
(df < 1000).all(axis=1)
````

### Filtrar por lista de valores: `isin()`

Cuando queremos seleccionar filas cuyo valor pertenece a un conjunto específico de opciones, `isin()` es más limpio que encadenar varios `==` con `|`:


````python
paises_latam = ["Chile", "Colombia", "Brazil", "Mexico"]

df.loc[df.index.isin(paises_latam)]
````

### Consultas con `query()`

`query()` es una alternativa más legible a las máscaras booleanas cuando trabajamos con condiciones sobre columnas. La condición se expresa como un **string**:


````python
# Equivalente a: df.loc[(df["Water quality"] >= 80) & (df["Air pollution"] < 15)]
df.query("`Water quality` >= 80 and `Air pollution` < 15")
````

### El parámetro `inplace`

Muchos métodos de pandas aceptan el argumento `inplace=True`, que modifica el DataFrame directamente sin necesidad de reasignar:

```python
df_ambiental.sort_values("Water quality", ascending=False, inplace=True)  # modifica df_ambiental

```

> **Nota 📖**: En la práctica se prefiere la forma de reasignación (`df = df.method(...)`) porque es más explícita, compatible con _method chaining_ y no presenta problemas con Copy-on-Write en pandas 2.0+. El uso de `inplace=True` está **desaconsejado** en código moderno.

### 7.5 Aplicar funciones a filas o columnas: `apply()`

`apply()` permite aplicar una función arbitraria a cada fila (`axis=1`) o columna (`axis=0`) de un DataFrame, o a cada elemento de una Serie.

> **⚠️ Advertencia de rendimiento**: `apply()` ejecuta la función en un bucle Python puro, por lo que es **significativamente más lento** que las operaciones vectorizadas nativas de pandas (como la aritmética entre columnas o `.str.method()`). Úsalo solo cuando no exista una alternativa vectorizada.


````python
def clasificar_contaminacion(valor):
    if valor < 10:
        return "Bajo"
    elif valor < 20:
        return "Moderado"
    else:
        return "Alto"


df_ambiental["Air pollution"].apply(clasificar_contaminacion)
````
